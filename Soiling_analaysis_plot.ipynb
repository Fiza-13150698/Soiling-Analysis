```python
#Simple mock representation of the data plotting done during soiling analysis
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
%matplotlib inline
df=pd.read_excel("PV_Soiling_Analysis.xlsx")
coverage=df["Cell Coverage (%)"]
power=df["Power (W)"]
ax=df.plot(x="Cell Coverage (%)",y="Power (W)")
initial_power = 380
final_power_threshold = initial_power * (1 - 0.3333)  # 33.33% reduction
ax.axhline(y=final_power_threshold, color='red', linestyle='--', linewidth=3,
          label=f'{final_power_threshold:.1f}W (33% â†“)')
ax.annotate('33.3% Power Loss', xy=(0, final_power_threshold),
            xytext=(5, final_power_threshold+15))
plt.ylim(0,400)
plt.ylabel("Power (W)")
plt.title("Power Performance-Soiling Anlaysis")
plt.show()

```
